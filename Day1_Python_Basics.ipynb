5. Functions

Write a function that:

accepts a list of numbers

returns mean, max, and min

```python
def summarize(numbers):
  meann= sum(numbers)/ len(numbers)
  maxn= max(numbers)
  minn= min(numbers)
  return meann, maxn, minn

```

Try it on your scores list

```python
sl= summarize(scores)
print(sl)
```

Vowel Counter

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    return sum(1 for char in text if char in vowels)
```

Duplicate Remover Tool

```python
def remove_duplicates(my_list):
    unique_items = []
    for item in my_list:
        if item not in unique_items:
            unique_items.append(item)
    return unique_items
```
